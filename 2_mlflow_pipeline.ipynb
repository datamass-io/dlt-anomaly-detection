```
# Lets load data from the training dataset
trainDF = spark.sql("SELECT * FROM hive_metastore.default.silver_kddcup_data_train")
display(trainDF)
```

```
# We also load the prepared test dataset
testDF = spark.sql("SELECT * FROM hive_metastore.default.silver_kddcup_data_test")
display(testDF)
```

```
from pyspark.sql.functions import col
display(trainDF.select('type').groupBy('type').count().orderBy(col('count').desc()))
display(trainDF.select('protocol_type').groupBy('protocol_type').count().orderBy(col('count').desc()))
display(trainDF.select('service').groupBy('service').count().orderBy(col('count').desc()))
display(trainDF.select('flag').groupBy('flag').count().orderBy(col('count').desc()))
```

```
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
import mlflow
from mlflow.models.signature import infer_signature

with mlflow.start_run(run_name='custom_random_forest'):
    stringIndexerProtocol = StringIndexer(inputCol="protocol_type", outputCol="indexed_protocol_type", stringOrderType="frequencyDesc")
    stringIndexerService = StringIndexer(inputCol="service", outputCol="indexed_service", stringOrderType="frequencyDesc")
    stringIndexerFlag = StringIndexer(inputCol="flag", outputCol="indexed_flag", stringOrderType="frequencyDesc")

    indexingPipeline = Pipeline(stages=[stringIndexerProtocol, stringIndexerService, stringIndexerFlag])

    # RF doesn't need OHE
    oheProtocol = OneHotEncoder(inputCols=['indexed_protocol_type'], outputCols=['ohe_protocol_type'])
    oheService = OneHotEncoder(inputCols=['indexed_service'], outputCols=['ohe_service'])
    oheFlag = OneHotEncoder(inputCols=['indexed_flag'], outputCols=['ohe_flag'])

    ohePipeline = Pipeline(stages=[oheProtocol, oheService, oheFlag])
    
    assembler = VectorAssembler(inputCols=[col for col in trainDF.columns if col not in ['type', 'packet_id', 'flag', 'protocol_type', 'service', 'difficulty'] + ['ohe_protocol_type', 'ohe_service', 'ohe_flag']], outputCol='featureVector')
    
    # but it needs Indexed labels
    stringIndexerForType = StringIndexer(inputCol="type", outputCol="typeIndexed", stringOrderType="frequencyDesc", handleInvalid="keep")

    dataPipeline = Pipeline(stages=[indexingPipeline, assembler, stringIndexerForType])

    # Training
    maxDepth = 6 
    numTrees = 24
    maxBins = 72
    
    mlflow.log_param('maxDepth', maxDepth)
    mlflow.log_param('numTrees', numTrees)
    mlflow.log_param('maxBins', maxBins)

    randomForest = RandomForestClassifier(featuresCol="featureVector", labelCol='typeIndexed', maxBins=maxBins, maxDepth=maxDepth, numTrees=numTrees)

    fullPipeline = Pipeline(stages=[dataPipeline, randomForest])
    fullPipelineModel = fullPipeline.fit(trainDF)
    
    # log the final full pipeline model
    mlflow.spark.log_model(spark_model=fullPipelineModel, artifact_path="FullPipeline")    

    # Testing
    predictions = fullPipelineModel.transform(testDF)
    
    from pyspark.ml.evaluation import MulticlassClassificationEvaluator
    evaluator = MulticlassClassificationEvaluator(predictionCol='prediction', labelCol='typeIndexed', metricName='accuracy')
    accuracy = evaluator.evaluate(predictions)

    evaluator = MulticlassClassificationEvaluator(predictionCol='prediction', labelCol='typeIndexed', metricName='f1')
    f1 = evaluator.evaluate(predictions)
    
    evaluator = MulticlassClassificationEvaluator(predictionCol='prediction', labelCol='typeIndexed', metricName='precisionByLabel')
    precisionByLabel = evaluator.evaluate(predictions)
    
    evaluator = MulticlassClassificationEvaluator(predictionCol='prediction', labelCol='typeIndexed', metricName='recallByLabel')
    recallByLabel = evaluator.evaluate(predictions)

    evaluator = MulticlassClassificationEvaluator(predictionCol='prediction', labelCol='typeIndexed', metricName='fMeasureByLabel')
    fMeasureByLabel = evaluator.evaluate(predictions)

    # log the metrics
    mlflow.log_metric('accuracy', accuracy)
    mlflow.log_metric('f1', f1)
    mlflow.log_metric('precisionByLabel', precisionByLabel)
    mlflow.log_metric('recallByLabel', recallByLabel)
    mlflow.log_metric('fMeasureByLabel', fMeasureByLabel)
    
```
